```python
wf = pd.read_csv('weather.csv')
wf.info()
```

```python
wf.head()
```

```python
wf = wf.replace('M',nan)
wf = wf.convert_objects(convert_numeric=True)
wf['date'] = pd.to_datetime(wf['date'])
```

```python
wf.info()
```

```python
wf['codesum'].unique()
```

```python
ar = wf['codesum'].unique()
cdsm = set()
for a in ar:
    b = a.split(' ')
    for c in b:
        d = ' '
        if len(c) == 4:
            c,d = c[0:2],c[2:4]
        elif len(c) == 3:
            c = c[0:2]
        cdsm.add(c)
        if d != ' ':
            cdsm.add(d)
cdsm.remove('')           
print cdsm            
```

```python
for k in cdsm:
    wf[k] = pd.Series(0)
    wf[k].fillna(0)
```

```python
for i, a in enumerate(wf['codesum']):
    a = a.strip()
    b = a.split(' ')
    for c in b:
        if len(c) == 4:
            wf[c[0:2]][i] = 1
            wf[c[2:4]][i] = 1
        elif len(c) == 3:
            wf[c[0:2]][i] = 2
        elif len(c) == 2:
            wf[c][i] == 1
```
