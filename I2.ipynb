```python
import pandas as pd
import numpy as np

wf = pd.read_csv('weather.csv')
wf = wf.replace('M',nan)
wf = wf.convert_objects(convert_numeric=True)
wf['date'] = pd.to_datetime(wf['date'])
ar = wf['codesum'].unique()
cdsm = set()
for a in ar:
    b = a.split(' ')
    for c in b:
        d = ' '
        if len(c) == 4:
            c,d = c[0:2],c[2:4]
        elif len(c) == 3:
            c = c[0:2]
        cdsm.add(c)
        if d != ' ':
            cdsm.add(d)
cdsm.remove('')
for k in cdsm:
    wf[k] = pd.Series(0)
    wf[k] = wf[k].fillna(0)

for i, a in enumerate(wf['codesum']):
    a = a.strip()
    b = a.split(' ')
    for c in b:
        if len(c) == 4:
            wf[c[0:2]][i] = 1
            wf[c[2:4]][i] = 1
        elif len(c) == 3:
            wf[c[0:2]][i] = 2
        elif len(c) == 2:
            wf[c][i] == 1
wf = wf.drop(['codesum'], axis=1)
```

```python
cols = wf.columns.tolist()
for col in cols:
    if col == 'station_nbr' or col == 'date':
        continue
    wf[col] = wf[col].astype(np.float32)
wf.info()
```
